```python
import pandas as pd

from statx_helpers import save_all_datasets

input_files = {
    "all_drugs": "../outputs/train_all_drugs_merged.csv",
    "all_opioids": "../outputs/train_all_opioids_merged.csv",
    "all_stimulants": "../outputs/train_all_stimulants_merged.csv",
}

all_summaries = []

for category, path in input_files.items():
    print(f"\nProcessing: {category}")
    df = pd.read_csv(path)

    summary, train_periods, val_periods = save_all_datasets(
        df,
        output_dir="../outputs",
        prefix=category,
        val_size=0.2,
        random_state=42,
    )
    summary.insert(0, "prediction_type", category)
    all_summaries.append(summary)

    print("Raw shape:", df.shape)
    print("Train periods:", len(train_periods), "| Validation periods:", len(val_periods))

summary_df = pd.concat(all_summaries, ignore_index=True)
display(summary_df)
print("All period-split datasets saved successfully.")
```
