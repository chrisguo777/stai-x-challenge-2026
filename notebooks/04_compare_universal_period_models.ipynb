```python
import pandas as pd

from statx_helpers import DATASET_SPECS, TARGET_COL, cross_val_by_period

# 5-fold GroupKFold by period (cross_val_by_period): every period is evaluated
# once and no period spans train/test, so the score is lower-variance than a
# single 80/20 holdout. Weather imputation is re-fit inside each fold (no leak).
raw_df = pd.read_csv("../outputs/train_universal_merged.csv")
print("Rows:", len(raw_df.dropna(subset=[TARGET_COL])),
      "| Periods:", raw_df["period_id"].nunique(), "| Folds: 5")

results = []
for dataset_name, spec in DATASET_SPECS.items():
    print(f"\n==================== {dataset_name} ====================")
    results.extend(cross_val_by_period(raw_df, spec["method"], dataset_name, n_splits=5))

results_df = pd.DataFrame(results).sort_values("RMSE").reset_index(drop=True)
results_df
```

```python
results_df.to_csv("../outputs/universal_period_model_comparison.csv", index=False)
print("Saved: ../outputs/universal_period_model_comparison.csv")
```
