```python
import pandas as pd

from statx_helpers import (
    DATASET_SPECS,
    TARGET_COL,
    fit_and_score_dataset,
    make_dataset,
    split_by_period,
)

raw_df = pd.read_csv("../outputs/train_universal_merged.csv")
train_raw, val_raw, train_periods, val_periods = split_by_period(
    raw_df.dropna(subset=[TARGET_COL]),
    val_size=0.2,
    random_state=42,
)

print("Train periods:", len(train_periods), "| Validation periods:", len(val_periods))
print("Train rows:", len(train_raw), "| Validation rows:", len(val_raw))

results = []
for dataset_name, spec in DATASET_SPECS.items():
    print(f"\n==================== {dataset_name} ====================")
    train_df = make_dataset(train_raw, spec["method"])
    val_df = make_dataset(val_raw, spec["method"], reference=train_raw)
    results.extend(fit_and_score_dataset(train_df, val_df, dataset_name))

results_df = pd.DataFrame(results).sort_values("RMSE").reset_index(drop=True)
results_df
```
